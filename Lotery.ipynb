```python
import math
import pandas as pd
```

```python
n = 60 # quantidade de números disponíveis para aposta
k = 6  # máximo de acertos
p = 7  # quantidade de números apostados

res = pd.DataFrame()

for p in range(k, 16):
    print(f'Aposta com {p} números:')
    for i in range(k, -1, -1):    
        num = math.comb(p,i)*math.comb(n-p,k-i)
        den = math.comb(n,k)
        prob = num/den
        print(f'\tAcertar {i} números: 1 em {round(den/num)} ({round(100*num/den,2)} %)')
        res = pd.concat([res, pd.DataFrame({
            'total_options': [n],
            'max': [k],
            'bet_amount': [p],
            'target': [i],
            'prob': [num/den],
            'one_in': round(den/num),
            'perc': round(100*num/den,2)
        })])
res = res.reset_index(drop=True)
```
